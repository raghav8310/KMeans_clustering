```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("C:/Users/DITU/Downloads/datasets-master/titanic.csv")
df
```

```python
df=df.dropna()
x=df["Age"]
y=df["Fare"]
```

```python
data= list(zip(x.values,y.values))
(data)
```

```python
model = KMeans(n_clusters = 2)
model.fit(data)
```

```python
plt.scatter(x,y,c=model.labels_)
plt.show
```
